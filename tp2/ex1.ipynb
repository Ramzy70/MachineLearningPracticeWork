```python
from sklearn.datasets import make_regression
import numpy as np

# Generate synthetic data for regression
x, y = make_regression(n_samples=100, n_features=1, noise=15)

# Adding intercept term to x
x = np.hstack((x, np.ones(x.shape)))

# Reshape y to ensure it's a column vector
y = y.reshape(100, 1)

# Initialize theta (parameters)
theta = np.random.rand(2, 1)

# Define the model function
def model(x, theta):
    return x.dot(theta)

# Calculate y_hat (predicted y values)
y_hat = model(x, theta)
```

```python
import matplotlib.pyplot as plt

# Plot the original data points
plt.scatter(x[:, 0], y, color='blue', label='Original data')

# Plot the regression line
plt.plot(x[:, 0], y_hat, color='red', label='Regression line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()

# Show plot
plt.show()
```

```python
def compute_cost(X, y, theta):
    m = len(y)
    predictions = np.dot(X, theta)
    error = predictions - y
    cost = (1/(2*m)) * np.sum(error ** 2)
    return cost

# Calculate the cost
cost = compute_cost(x, y, theta)
print("Initial cost:", cost)
```

```python
def grad (X,y,theta):
    m = len(y)
    return ((1/m)*X.T.dot(model(X,theta)-y))

# Calculate the gradient
g = grad(x, y, theta)
print("Initial gradient:", g)
```

```python
def gradient_decent (x, y , theta  ) :
    return theta - x.dot(grad(x, y, theta))
gd = gradient_decent(x, y, theta )
print("gradient decent:", gd)
```
