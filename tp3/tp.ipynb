```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Define the dataset
class0 = np.array([[1, 9], [5, 5], [1, 1]])
class1 = np.array([[8, 5], [13, 1], [13, 9]])
X = np.vstack((class0, class1))
y = np.concatenate((np.zeros(class0.shape[0]), np.ones(class1.shape[0])))

# Train a linear SVM classifier with a large C value
svm_classifier = SVC(kernel='linear', C=1e6)
svm_classifier.fit(X, y)

# Extract support vectors and coefficients of the decision boundary
support_vectors = svm_classifier.support_vectors_
coef = svm_classifier.coef_[0]
intercept = svm_classifier.intercept_

# Plot the data points
plt.scatter(class0[:, 0], class0[:, 1], c='blue', label='Class0')
plt.scatter(class1[:, 0], class1[:, 1], c='red', label='Class1')

# Plot the support vectors
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, facecolors='none', edgecolors='black', label='Support Vectors')

# Plot the decision boundary
x_vals = np.linspace(0, 15, 100)
y_vals = -(coef[0] * x_vals + intercept) / coef[1]
plt.plot(x_vals, y_vals, '--', color='green', label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear SVM Decision Boundary with Maximum Margin')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Define the dataset
class0 = np.array([[2, 2], [3, 3], [4, 4], [5, 5], [4, 6], [3, 7], [4, 8], [5, 9], [6, 10]])
class1 = np.array([[6, 2], [7, 3], [8, 4], [9, 5], [8, 6], [7, 7], [7, 8], [7, 9], [8, 10]])
X = np.vstack((class0, class1))
y = np.concatenate((np.zeros(class0.shape[0]), np.ones(class1.shape[0])))

# Train a linear SVM classifier with a high cost of misclassifying training data
svm_classifier = SVC(kernel='linear', C=1e6)
svm_classifier.fit(X, y)

# Extract support vectors and coefficients of the decision boundary
support_vectors = svm_classifier.support_vectors_
coef = svm_classifier.coef_[0]
intercept = svm_classifier.intercept_

# Plot the data points
plt.scatter(class0[:, 0], class0[:, 1], c='blue', label='Class0')
plt.scatter(class1[:, 0], class1[:, 1], c='red', label='Class1')

# Plot the support vectors
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, facecolors='none', edgecolors='black', label='Support Vectors')

# Plot the decision boundary
x_vals = np.linspace(2, 9, 100)
y_vals = -(coef[0] * x_vals + intercept) / coef[1]
plt.plot(x_vals, y_vals, '--', color='green', label='Decision Boundary')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear SVM Decision Boundary with High Cost')
plt.legend()
plt.grid(True)
plt.show()

# Calculate error rate
error_rate = 1 - svm_classifier.score(X, y)
print("Error rate of the linear SVM on the training dataset:", error_rate)
```
